## Grundlagen

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set()

# Beispieldaten einlesen, Zufallsstichprobe von n=100
df = pd.read_csv("../data/Library_Usage.csv").sample(n=100)
```

## Streudiagramme mit relplot

```python
# Erster Plot
sns.relplot(x='Total Checkouts',y='Total Renewals',data=df)
```

```python
# Plotgre: Hhe in inches, Breite = aspect*height
sns.relplot(x='Total Checkouts',y='Total Renewals',data=df,height=5,aspect=3)
```

```python
# Speichern
plt.savefig('../tutorial.png',dpi=150)
```

```python
# Weitere Eigenschaften
sns.relplot(x='Total Checkouts',y='Total Renewals',hue='Provided Email Address',style='Age Range',size='Year Patron Registered', data=df)
```

```python
# Plot-Gitter mit den Argumenten row und col
sns.relplot(x='Total Checkouts',y='Total Renewals',size='Year Patron Registered',row='Provided Email Address',col='Age Range',data=df)
```

```python
# Farbpaletten mit dem Argument palette fr den Parameter in hue, Name der Palette und Anzahl der Farben
palette =sns.color_palette('Accent',10)
sns.relplot(x='Total Checkouts',y='Total Renewals',hue='Age Range',palette=palette,data=df)
```

## Visualisierung mit catplot

```python
df = pd.read_csv("../data/Library_Usage.csv")
```

```python
# Nominale und ordinale Variablen
# univariate Hufigkeiten und Bivariate Kreuztabellen mit Balkendiagrammen
sns.catplot(y='Year Patron Registered',data=df,kind='count',color='steelblue')
```

```python
sns.catplot(y='Age Range',hue='Provided Email Address',data=df,kind='count')
```

```python
sns.catplot(x='Patron Type Definition',data=df,kind='count',col='Year Patron Registered',col_wrap=4)
```
