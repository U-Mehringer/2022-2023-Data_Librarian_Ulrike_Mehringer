## Matplotlib example

```python
# Matplotlib importieren
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Beispieldaten
x = range(1,10)
```

```python
# make simple scatter plot of data
plt.plot(x,x,c='blue', linestyle="", marker="v")
```

```python
plt.plot(x,x,c='magenta',linestyle="-")
```
