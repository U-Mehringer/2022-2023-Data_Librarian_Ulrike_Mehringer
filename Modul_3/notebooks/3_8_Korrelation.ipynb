```python
# Anscombe-Quartett

# 1. Einlesen
import pandas as pd
df = pd.read_csv("../data/anscombe.csv")
df.head()
```

```python
# 2. Mittelwert, Median, Standardabweichungen der Spalten
for z in df.columns:
    print(z,": Mittelwert:",df[z].mean(),"- Median:",df[z].median(),"- Standardabw.:",df[z].std())
```

```python
# 3. Korrelation zweier zusammenh√§ngender Spalten
# nummer = ["1","2","3","4"]
# for n in nummer:
#    x = "x"+n
#    y = "y"+n
#    anscombe = df[x].corr(df[y])
#    df["corr"+n] = anscombe
#    print("Zeile",i,"x/y",n,": Korrelation:",anscombe)

for i in range (1,5):
    anscombe = df[f'x{i}'].corr(df[f'y{i}'])
    print("x/y",i,", Korrelation: ",anscombe)
 
```
