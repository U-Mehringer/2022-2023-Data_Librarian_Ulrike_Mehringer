```python
import pandas as pd
import urllib.request
import json
```

# Weg 1

```python
with urllib.request.urlopen("https://www.bibsonomy.org/json/search/botanik?items=1000") as url:
    data = json.load(url)
    print(data)
```

```python
type(data)
```

```python
data.keys()
```

```python
botanik_df = pd.DataFrame(data["items"])
```

```python
botanik_df
```

# Weg 2

```python
data = urllib.request.urlopen("https://www.bibsonomy.org/json/search/botanik?items=1000").read()
output = json.loads(data)
print(output)
```

```python
type(output)
```

```python
output.keys()
```

```python
botanik_df = pd.DataFrame(output["items"])
```
