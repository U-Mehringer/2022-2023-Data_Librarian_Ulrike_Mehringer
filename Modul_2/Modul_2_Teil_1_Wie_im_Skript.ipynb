```python
import urllib *
# import urllib.parse ... request
import json
import requests
import pandas as pd
```

# Methode 1

```python
url = "https://www.bibsonomy.org/json/search/Botanik?items=1000&duplicates=merged"

output = urllib.request.urlopen(url)
data = output.read().decode('utf-8') # data ist vom Typ string
botanik = json.loads(data)
```

```python
type(botanik)
```

```python
print(botanik)
```
